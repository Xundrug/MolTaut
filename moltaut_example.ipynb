```python
from rdkit import Chem
import pandas as pd
from predict_tautomer import get_taut_data
import mols2grid
```

```python
def construct_data(out_data):
    new_data = []
    for info in out_data:
        new_data.append([Chem.MolFromSmiles(info["tsmi"]), info["score"], info["label"]])
    df_data = pd.DataFrame(new_data)
    df_data.columns = ["tmol", "score", "label"]
    return df_data
```

```python
smi = "Cc1c2c([nH]n1)OC(=C([C@@]2(c3cc(cc(c3)N4CCCC4)C(F)(F)F)C(C)C)C#N)N"
energy_cutoff = 2.76
cutmol = True
num_confs = 32
ph = 7.0
tph = 1.0
```

```python
data = get_taut_data(smi, cutmol, num_confs, energy_cutoff, ph, tph)
```

```python
df = construct_data(data)
```

```python
mols2grid.display(df, mol_col="tmol", subset=["img", "score", "label"], n_cols=3)
```
